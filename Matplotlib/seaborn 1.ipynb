```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('Iris.csv')
data
```

```python
sns.boxplot(x = 'Species',y = 'SepalLengthCm',data = data)
```

```python
sns.countplot(data['Species'])
```

```python
sns.heatmap(data.corr(), annot = True, cmap = 'Accent')
```

```python
sns.violinplot(data['Species'], data['SepalLengthCm'])
```

```python
plt.scatter(x = data['Species'],y = data['SepalLengthCm'])
plt.xlabel('S')
plt.ylabel('SeapalLengthCm')
```

```python
plt.scatter(x = data['Species'],y = data['SepalLengthCm'], color = 'green')
```

```python
sns.distplot(data['SepalWidthCm'])
```

```python
plt.bar(data['Species'], data['SepalLengthCm'])
```

```python
data.groupby('Species')['SepalLengthCm'].mean().plot(kind = 'bar')
```
