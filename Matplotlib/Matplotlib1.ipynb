```python
import matplotlib.pyplot as plt
print(matplotlib.__version__)
#Now the version available as 3.5.1
import numpy as np
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y,'o')
plt.xlabel('x')
plt.ylabel('y')
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker = '*')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, '-.')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker = 'o', color = 'green')
plt.show()
```
