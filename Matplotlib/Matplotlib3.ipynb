```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [2,3,4]

plt.plot(x, y, marker = 'o')
plt.grid()
```

```python
#subplots
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, marker = 'o')
plt.grid()
plt.title('x,y')

plt.subplot(2,1,2)
plt.plot(y,x, marker = 'o')
plt.title('y,x')
```

```python
x = np.random.randint(100, size = (100))
plt.scatter(x,x, c=x, s=x, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
```

```python
x = [10,20,30,40]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(x, labels = mylabels)
plt.title('fruits')
plt.show()
```

```python
myexplode = [0.2, 0, 0, 0]

plt.pie(x, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
myexplode = [0.2, 0, 0, 0]

plt.pie(x, labels = mylabels, explode = myexplode)
plt.legend(loc = 'upper left')
plt.show() 
```
