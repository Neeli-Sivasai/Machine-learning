```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker = 'o', color = 'green', ms = 10)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker = 'o', color = 'green', ms = 15, mfc = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, marker = 'o', color = 'green', ms = 15, mfc = 'b', mec = 'b' )
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints, linewidth = 5)
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.grid()
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```
